```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))

plt.plot(data['Date'].sort_index(ascending = False), data['Close'].sort_index(ascending = False))
plt.title('Tesla Stock Close Price in USd')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks([0, 150, 300, 450, 603, 753])
plt.show()
```
